## 合并

```python
import re
import csv
path = r"C:\Users\Administrator\Desktop\source.txt"
with open(path,'r',encoding = 'utf-8') as f:
    source = f.read()

username_list = re.findall('username="(.*?)"',source,re.S)
content = re.findall('d_post_content j_d_post_content " style="display:;">(.*?)<',source,re.S)
replytime = re.findall('span class="tail-info">(2025.*?)<',source, re.S)

result_list = []
for i in range(len(username_list)):
    result = {
        'username':username_list[i],
        'content':content[i],
        'reply_time':replytime[i]
    }
    result_list.append(result)

with open('tieba.csv','w',encoding = 'utf-8') as f:
    writer = csv.DictWriter(f,fieldnames = ['username','content','reply_time'])
    writer.writeheader()
    writer.writerows(result_list)
```
